## <u>Selecting Series data</u>
<p>The commonly used indexing operators are loc and iloc. <b>loc</b> is used to fetch label locators and <b>iloc</b> is used to fetch index locators</p>

```python
import numpy as np
import pandas as pd
```

```python
college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
```

```python
college.head()
```

```python
#Selecting a single column as a Series:
city = college['CITY']
```

```python
city.head()
```

```python
#Indexing starts from zero. To select the fourth row:
city.iloc[3]
```

```python
#Selection of several integer locations:
city.iloc[[10, 20, 30]]
```

```python
#Selecting an equally spaced partitions of data:
#It uses the start, stop and steps syntax
city.iloc[4:50:10]
```

```python
#Selecting labeled data by index:
city.loc['Heritage Christian University']
```

```python
#Selecting 4 random rows using the lists format:
np.random.seed(1)
labels = list(np.random.choice(city.index, 4))
labels
```

```python
city.loc[labels]
```

```python
#Equally spaced data using equal step sizes:
city.loc['Alabama State University': 'Reid State Technical College':10]
```

```python
#To get an output that also includes the index that the filter had as its output:
city.iloc[[3]]
```

```python
#This was the initial output that was being obtained:
city.iloc[3]
```

```python
#If the start is after the stop in the steps format, an empty Series is returned without an exception raised:
city.loc['Reid State Technical College': 'Alabama State University' : 10]
```

```python
city.loc['Alabama State University' : 'Reid State Technical College' : 10]
```

## <u>Selecting data with both integers and labels</u>
<p>In previous Pandas versions, the .ix indexer had the ability to select both integer and label location but it was very
ambiguous and was a source of confusion for many Pandas users. The indexer is currently depreciated and should be avoided. It 
was previously possible to write selection code such as <b><i>college.ix[:5,'UGDS_WHITE':'UGDS_UNKN'].</i></b> This is now impossible when loc or iloc are used. We will now get into finding the integer location of the columns then use iloc to complete the selection.</p>

```python
college = pd.read_csv('data/college.csv', index_col = 'INSTNM')
```

```python
college.head()
```

```python
#Use the get_loc index method to get the index location of the desired columns:
col_start = college.columns.get_loc('UGDS_WHITE')
col_end = college.columns.get_loc('UGDS_UNKN') + 1
col_start, col_end
```

```python
#Use col_start and col_end to select the first five rows of columns by integer location using iloc:
college.iloc[:5, col_start : col_end]
```

```python
#Selecting an entire row at that position:
college.iloc[2]
```

```python
#Alternatively:
college.loc['University of Alaska Anchorage']
```

```python
#Selection of a disjointed set of rows:
college.iloc[[60, 99, 3]]
```

```python
#Alternatively:
labels = ['University of Alaska Anchorage',
          'International Academy of Hair Design',
          'University of Alabama in Huntsville']
```

```python
#You can also get the list of indexes without copying and pasting:
college.iloc[[60, 99, 3]].index.tolist()
```

```python
college.loc[labels]
```

```python
#Slicing can also be done as follows:
college.iloc[99:102]
```

```python
college.iloc[102]
```

```python
#Slice notation works with the .loc indexer and it is inclusive of the last label:
start = 'International Academy of Hair Design'
stop = 'Mesa Community College'
college.loc[start:stop]
```

```python
#A replacement of the get_index method is this one:
row_start = college.index[10]
row_end = college.index[15]
college.loc[row_start : row_end, 'UGDS_WHITE':'UGDS_UNKN']
```

```python
#Chaining indexers is a bad idea but it can work:
college.iloc[10:16].loc[:, 'UGDS_WHITE':'UGDS_UNKN']
```

## <u>Speeding up scalar selection</u>
<p>There are indexers <b>iat</b> and <b>at</b> which achieve the same thing at faster speeds. They only accept scalar values. They save several microseconds per loop if they may be used in a program</p>

```python
cn = 'Texas A & M University-College Station'
```

```python
college.loc[cn, 'UGDS_WHITE']
```

```python
#The %timeit magic command can be used to compare the difference in speed:
%timeit college.loc[cn, 'UGDS_WHITE']
```

```python
%timeit college.at[cn, 'UGDS_WHITE']
```

```python
row_num = college.index.get_loc(cn)
row_num
```

```python
col_num = college.columns.get_loc('UGDS_WHITE')
col_num
```

```python
row_num, col_num
```

```python
%timeit college.iloc[row_num, col_num]
```

```python
%timeit college.iat[row_num, col_num]
```

```python
#.iat and .at work with Series as well. Pass them a scalar value and they will retur a scalar:
state = college['STABBR']
```

```python
state.iat[1000]
```

```python
state.at['Stanford University']
```

## <u>Slicing rows lazily</u>
<p>Rows can be sliced lazily without the use of loc or iloc. It does not work for columns, but just for DataFrame rows and Series. It cannot also be used to select rows and columns simultaneously</p>

```python
college[10 : 20 : 2]
```

```python
#The same slicing exists with Series:
city = college['CITY']
```

```python
city[10 : 20 : 2]
```

```python
#Series and DataFrames can slice by label with just the indexing operator:
start = 'Mesa Community College'
stop = 'Spokane Community College'
college[start:stop:1500]
```

```python
college.index.get_loc('Mesa Community College')
```

```python
college.index.get_loc('Spokane Community College')
```

```python
#Here is the same slice label with a Series:
city[start:stop:1500]
```

## <u>Slicing lexicographically</u>
<p>The loc indexer can also be used to select data based on the order of the values in the index. It works with the slice notation only if the index is sorted:</p>

```python
college = college.sort_index()
```

```python
college.head()
```

```python
#This method is case-sensitive and is inclusive of the last index
#Python sorts capital letters before small letters and integers before capital letters
college.loc['Sp' : 'Su']
```

```python
#It also works when the index is sorted in the opposite direction:
college = college.sort_index(ascending = False) 
```

```python
#Determining the direction which the index is sorted:
college.index.is_monotonic_decreasing
```

```python
college.index.is_monotonic_increasing
```

```python
college.loc['E' : 'B']
```

```python
college.loc['B' : 'E']
```
